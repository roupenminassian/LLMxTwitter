```python
tornadoDf = pd.read_csv('/content/2011Tornado_Summary.csv')
sandyDf = pd.read_csv('/content/2012Sandy_Summary.csv')
floodsDf = pd.read_csv('/content/2013Floods_Summary.csv')
blizzardDf = pd.read_csv('/content/2016Blizzard_Summary.csv')
matthewDf = pd.read_csv('/content/2016Matthew_Summary.csv')
hurricaneDf = pd.read_csv('/content/2017Hurricane_Summary.csv')
michaelDf = pd.read_csv('/content/2018Michael_Summary.csv')
wildfireDf = pd.read_csv('/content/2018Wildfires_Summary.csv')
dorianDf = pd.read_csv('/content/2019Dorian_Summary.csv')
```

```python
naturalDisasterDf = pd.concat([tornadoDf,sandyDf,floodsDf,blizzardDf,matthewDf,hurricaneDf,michaelDf,wildfireDf,dorianDf]).reset_index().drop(columns = 'index')
```

```python
from mumin import MuminDataset
dataset = MuminDataset(bearer_token, size='large')
dataset.compile()
```

```python
tweet_df = dataset.nodes['tweet']
tweet_df.dropna(inplace=True)
tweet_df.head()
```

```python
claim_df = dataset.nodes['claim']
claim_df.head(3)
```

```python
discusses_df = dataset.rels[('tweet', 'discusses', 'claim')]
discusses_df.head()
```

```python
tweet_claim_df = (tweet_df.merge(discusses_df, left_index=True, right_on='src')
                          .merge(claim_df, left_on='tgt', right_index=True)
                          .reset_index(drop=True))
tweet_claim_df.head(3)
```

```python
tweet_claim_df = tweet_claim_df[tweet_claim_df['lang'] == 'en'][['text','label']].reset_index().drop(columns=['index'])
```

```python
tweet_claim_df['label'] = tweet_claim_df['label'].map({'misinformation':1, 'factual':0})
```

##Concatenate Datasets

```python
train = pd.concat([fakeNewsDf,disasterDf,naturalDisasterDf,tweet_claim_df]).reset_index().drop(columns = 'index')
```

```python
import matplotlib.pyplot as plt

# Convert the labels in the 'label' column to strings
train['label'] = train['label'].astype(str)

# Group the labels and calculate their proportions
label_proportions = train.groupby('label')['label'].count() / len(train)

# Create a bar chart of the label proportions
fig, ax = plt.subplots()
ax.bar(label_proportions.index, label_proportions.values)

# Set labels and title for the chart
ax.set_ylabel('Proportion')
ax.set_xlabel('Labels')
ax.set_title('Proportions of Labels')

# Display the chart
plt.show()

#102,000
```

##Create Train and Test
