```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
pathname = '/content/drive/MyDrive/Twitter Analysis Project (DPA)/Twitter Data/'
```

```python
dfArm = pd.read_csv(pathname + 'finalarm.csv')
dfAz = pd.read_csv(pathname + 'finalaz.csv')
```

```python
def preprocessing(df):

  column_name = 'processed_full_text'

  # Remove duplicates
  df.drop_duplicates(inplace=True)

  # Remove rows with missing values
  df.dropna(inplace=True)

  df['processed_full_text'] = df['full_text']

  # Convert text to lowercase
  df[f'{column_name}'] = df[f'{column_name}'].apply(lambda x: x.lower())

  # Remove URLs
  df[f'{column_name}'] = df[f'{column_name}'].str.replace('http\\S+|www.\\S+', '', case=False)

  # Remove mentions
  df[f'{column_name}'] = df[f'{column_name}'].str.replace('@\\S+', '', case=False)

  # Remove hashtags
  #df[f'{column_name}'] = df[f'{column_name}'].str.replace('#\\S+', '', case=False)

  # Remove punctuation
  df[f'{column_name}'] = df[f'{column_name}'].str.replace('[^\\w\\s]','')

  # Remove stopwords
  #stop_words = stopwords.words('english')
  #df[f'{column_name}'] = df[f'{column_name}'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))

  return df
```

##Pro-Armenian Dataset

```python
dfArm.drop(columns=['Unnamed: 0','hashtags'], inplace=True)
```

```python
dfArm.columns = ['id','created_at','full_text','user.name','user.screen_name','user.id','time with YVN/Baku tmz']
```

```python
dfArm = dfArm[['created_at','id','full_text','user.name','user.screen_name','user.id','time with YVN/Baku tmz']]
```

```python
dfArm = preprocessing(dfArm)
```

```python
dfArm.sort_values(by='created_at', inplace = True)
```
